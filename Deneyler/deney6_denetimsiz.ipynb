```python
from sklearn import datasets
iris_veri=datasets.load_iris()
iris=iris_veri.data
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
model.fit(iris)
KMeans(algorithm='auto', copy_x=True , init='k-means++', max_iter=300, n_clusters=3 , n_init=10 
       ,random_state=None, tol=0.0001 , verbose=0)
gruplar=model.predict(iris)
print
```

```python
merkezler=model.cluster_centers_
print(merkezler)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x=iris[:,2]
y=iris[:,3]
merkez_x=merkezler[:,2]
merkez_y=merkezler[:,3]
plt.scatter(x,y,c=gruplar)
plt.scatter(merkez_x,merkez_y,marker='P',s=100, c='r')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

iris_veri=datasets.load_iris()
iris=iris_veri.data

from sklearn.cluster import KMeans

obek=10

sonuc=[]

for k in np.arange(1,obek+1):
    model=KMeans(n_clusters=k)
    model.fit(iris)
    sonuc.append(model.inertia_)
    
km=pd.DataFrame({'k' : np.arange(1,obek+1),'sonuc' : sonuc})
plt.plot(km['k'],km['sonuc'],'bo-')
plt.xlabel('Öbek Sayısı')
plt.ylabel('Kareler toplamı')
plt.show()
```

```python
#PCA principal component analysis temel bilinen analizi
from sklearn.decomposition import PCA
from sklearn import datasets 
import matplotlib.pyplot as plt

iris_veri=datasets.load_iris()
iris=iris_veri.data[:,2:4]

model=PCA()
iris_yeni=model.fit_transform(iris)

x=iris_yeni[:,0]
y=iris_yeni[:,1]

mean=model.mean_
first_pc=model.components_[0,:]

fig, ax=plt.subplots(nrows=1 , ncols=2)
ax[0].set_ylim(-0.6,2.6)
ax[1].set_ylim(-0.6,2.6)
ax[0].scatter(iris[:,0],iris[:,1],alpha=0.4,color='blue')
ax[0].arrow(mean[0],mean[1],first_pc[0],first_pc[1],color='red',width=0)
ax[1].axis('equal')
ax[1].scatter(x,y, alpha=0.4,color='magenta')
ax[0].set_xlabel('Orijinal veri')
ax[1].set_xlabel('Dönüştürülmüş veri')
ax[1].axis('equal')
plt.show()
```
